```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sea
```

```python
# Carregando um dos datasets que vem com o Seaborn
dados = sea.load_dataset("tips")
```

```python
dados.head()
```

```python
# O método joinplot cria plot de 2 variáveis com gráficos bivariados e univariados
# o método utiliza duas váriaveis do dataset "total_bill" e "tip" e define no argumento kind como os
#dados serão mostrado no gráfico esse métod o apresenta a relação de duas váriaveis
sea.jointplot("total_bill", "tip", dados, kind = 'hex');
sea.jointplot("total_bill", "tip", dados, kind = 'scatter');
sea.jointplot("total_bill", "tip", dados, kind = 'reg');
sea.jointplot("total_bill", "tip", dados, kind = 'resid');
sea.jointplot("total_bill", "tip", dados, kind = 'kde');
```

```python
# O método lmplot() cria plot com dados e modelos de regressão. comparando com smoker yes e no
sea.lmplot("total_bill", "tip", dados, col = "smoker");
```

```python
# Construindo um dataframe com Pandas
df = pd.DataFrame()
```

```python
# Alimentando o Dataframe com valores aleatórios o método sample pega(coleta) uma amostra 
#dos dados gerados
df['a'] = random.sample(range(1, 100), 25)
df['b'] = random.sample(range(1, 100), 25)
```

```python
print(df.b)
```

```python
# Scatter Plot
sea.lmplot('a', 'b', data = df, fit_reg = True);
```

```python
# Density Plot
sea.kdeplot(df.b);
```

```python
sea.kdeplot(df.b, df.a);
```

```python
sea.distplot(df.a);
```

```python
# Histograma
plt.hist(df.a, alpha = .3)
sea.rugplot(df.a);
```

```python

# Box Plot
sea.boxplot([df.b, df.a]);
```

```python
# Violin Plot
sea.violinplot([df.a, df.b]);
```

```python
# Heatmap
sea.heatmap([df.b, df.a], annot = True, fmt = "d");
```

```python
# Clustermap
sea.clustermap(df);
```
